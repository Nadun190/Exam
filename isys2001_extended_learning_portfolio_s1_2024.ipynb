**Instructions:**

Design and implement an application that helps users create and manage a monthly
budget. The application should prompt the user to input their monthly income and
various expense categories (e.g., rent, groceries, utilities, entertainment).
Your design should:

    category.
2.  Calculate and display the total expenses and the remaining balance after all

3.  Allow the user to repeat the process with different inputs to plan different

Your design should demonstrate the first five steps of the development

```python
def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Please enter a positive number.")
            else:
                return value
        except ValueError:
            print("Invalid input. Please enter a valid number.")

def calculate_budget(income, expenses):
    total_expenses = sum(expenses.values())
    remaining_balance = income - total_expenses
    return total_expenses, remaining_balance

def main():
    print("Welcome to the Monthly Budget Planner")

    while True:
        print("\nEnter your financial details for this month:")

        income = get_positive_float("Enter your monthly income: $ ")

        expense_categories = ["Rent", "Groceries", "Utilities", "Entertainment", "Others"]
        expenses = {}

        print("\nEnter your expenses for each category:")
        for category in expense_categories:
            expenses[category] = get_positive_float(f"Enter {category} expenses: $ ")

        total_expenses, remaining_balance = calculate_budget(income, expenses)

        print("\n--- Budget Summary ---")
        print(f"Monthly Income: ${income:.2f}")

        for category, amount in expenses.items():
            print(f"{category}: ${amount:.2f}")

        print(f"Total Expenses: ${total_expenses:.2f}")
        print(f"Remaining Balance: ${remaining_balance:.2f}")

        repeat = input("\nDo you want to plan another budget? (yes/no): ").strip().lower()
        if repeat != 'yes':
            break

    print("\nThank you for using the Monthly Budget Planner!")

# Run the application
if __name__ == "__main__":
    main()
```

```python
# Function to get monthly income from the user
def get_monthly_income():
    while True:
        try:
            income = float(input("Enter your monthly income: "))
            if income >= 0:
                return income
            else:
                print("Income should be a positive number. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Function to get expense amounts for various categories
def get_expenses():
    expenses = {}
    categories = ['rent', 'groceries', 'utilities', 'entertainment']  # Example expense categories

    for category in categories:
        while True:
            try:
                expense = float(input(f"Enter amount spent on {category}: "))
                if expense >= 0:
                    expenses[category] = expense
                    break
                else:
                    print("Expense should be a positive number. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a valid number.")

    return expenses

# Function to calculate total expenses
def calculate_total_expenses(expenses):
    total_expenses = sum(expenses.values())
    return total_expenses

# Function to calculate remaining balance
def calculate_remaining_balance(income, total_expenses):
    remaining_balance = income - total_expenses
    return remaining_balance

# Main function to run the budget planner
def main():
    print("Welcome to the Monthly Budget Planner\n")

    # Get monthly income
    income = get_monthly_income()

    # Get expenses
    expenses = get_expenses()

    # Calculate total expenses
    total_expenses = calculate_total_expenses(expenses)

    # Calculate remaining balance
    remaining_balance = calculate_remaining_balance(income, total_expenses)

    # Display results
    print("\n===== Monthly Budget Summary =====")
    print(f"Total Income: ${income}")
    print(f"Total Expenses: ${total_expenses}")
    print(f"Remaining Balance: ${remaining_balance}")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
def schedule_volunteers(volunteers: list, shifts: list) -> dict:
    """
    Assigns volunteers to available shifts based on their preferences.

    Args:
    volunteers (list): List of volunteer names
    shifts (list): List of shift names

    Returns:
    dict: Dictionary with shift names as keys and assigned volunteer names as values

    Example:
    >>> schedule_volunteers(["Alice", "Bob"], ["Morning", "Afternoon"])
    {'Morning': 'Alice', 'Afternoon': 'Bob'}
    """

    assignments = {}

    for volunteer in volunteers:
        while True:
            try:
                preference_num = int(input(f"{volunteer}, enter your preferred shift number (1-{len(shifts)}): "))
                if preference_num < 1 or preference_num > len(shifts):
                    raise ValueError("Invalid shift number.")
                preference_index = preference_num - 1  # Convert to 0-based index
                assignments[shifts[preference_index]] = volunteer
                shifts.pop(preference_index)  # Remove assigned shift
                break
            except ValueError as e:
                print(e)
                print("Please enter a valid shift number.")

    return assignments

if __name__ == "__main__":
    # List of volunteers
    volunteers = ["Alice", "Bob"]
    # List of shifts
    shifts = ["Morning", "Afternoon"]

    # Call the schedule_volunteers function
    result = schedule_volunteers(volunteers, shifts)

    # Print the results
    print("Volunteer Assignments:")
    for shift, volunteer in result.items():
        print(f"{shift}: {volunteer}")
```

## Question 3: Code City Adventures Implementation

**Background:**

"Code City Adventures" is a text-based adventure game where players use Python
to solve challenges in a modern city setting. In the first level, "The
Neighbourhood Watch," the player helps organise a schedule for volunteer watch
shifts. In the second level "The Automated Cafe", the player helps a café
automate its order system using conditionals to handle menu choices and

**Instructions:**

Implement the first level (The Neighbourhood Watch) and the second level (The
Automated Cafe) of the "Code City Adventures" game in one or more Google Colab

```python
# Level 1: The Neighbourhood Watch

def collect_volunteers():
    """
    Collects volunteer names and their available days.

    Returns:
        dict: A dictionary with volunteer names as keys and available days as values.
    """
    volunteers = {}
    while True:
        name = input("Enter volunteer's name (or type 'done' to finish): ")
        if name.lower() == 'done':
            break
        days = input("Enter days available (comma-separated, e.g., Mon,Tue,Wed): ")
        volunteers[name] = days.split(',')
    return volunteers

def create_schedule(volunteers):
    """
    Creates a schedule based on volunteer availability.

    Args:
        volunteers (dict): A dictionary with volunteer names as keys and available days as values.

    Returns:
        dict: A schedule with days as keys and volunteers as values.
    """
    schedule = {day: [] for day in ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']}
    for name, days in volunteers.items():
        for day in days:
            if day in schedule:
                schedule[day].append(name)
    return schedule

def display_schedule(schedule):
    """
    Displays the schedule.

    Args:
        schedule (dict): A schedule with days as keys and volunteers as values.
    """
    print("\nNeighbourhood Watch Schedule:")
    for day, volunteers in schedule.items():
        print(f"{day}: {', '.join(volunteers) if volunteers else 'No volunteers'}")

def neighbourhood_watch():
    """
    Runs the Neighbourhood Watch level.
    """
    print("Welcome to The Neighbourhood Watch!\n")
    print("Help us organize a schedule for the volunteer watch shifts.\n")
    volunteers = collect_volunteers()
    schedule = create_schedule(volunteers)
    display_schedule(schedule)

# Level 2: The Automated Cafe

def display_menu():
    """
    Displays the cafe menu.
    """
    menu = {
        'coffee': 2.50,
        'tea': 2.00,
        'sandwich': 5.00,
        'salad': 4.50
    }
    print("\nCafe Menu:")
    for item, price in menu.items():
        print(f"{item.title()}: ${price:.2f}")
    return menu

def take_order(menu):
    """
    Takes the customer's order.

    Args:
        menu (dict): A dictionary with menu items and their prices.

    Returns:
        list: A list of ordered items.
    """
    order = []
    while True:
        item = input("Enter item to order (or type 'done' to finish): ").lower()
        if item == 'done':
            break
        if item in menu:
            order.append(item)
        else:
            print("Item not on the menu. Please choose again.")
    return order

def calculate_total(order, menu):
    """
    Calculates the total cost of the order.

    Args:
        order (list): A list of ordered items.
        menu (dict): A dictionary with menu items and their prices.

    Returns:
        float: The total cost of the order.
    """
    total = sum(menu[item] for item in order)
    return total

def automated_cafe():
    """
    Runs the Automated Cafe level.
    """
    print("Welcome to The Automated Cafe!\n")
    print("Help us automate our order system.\n")
    menu = display_menu()
    order = take_order(menu)
    total = calculate_total(order, menu)
    print(f"\nYour order: {', '.join(order)}")
    print(f"Total cost: ${total:.2f}")

# Run the levels
print("Code City Adventures\n")
print("Level 1: The Neighbourhood Watch")
neighbourhood_watch()

print("\nLevel 2: The Automated Cafe")
automated_cafe()
```
